```python
cmd = None

dict1 = {"Quantità": 30, "Prezzo di acquisto": 0.80, "Prezzo di vendita": 1.40}
dict2 = {"Quantità": 20, "Prezzo di acquisto": 3.10, "Prezzo di vendita": 5.20}
dict3 = {"Quantità": 40, "Prezzo di acquisto": 2.25, "Prezzo di vendita": 4.45}
dict4 = {"Quantità": 15, "Prezzo di acquisto": 2.50, "Prezzo di vendita": 5.10}
dict5 = {"Quantità": 35, "Prezzo di acquisto": 1.20, "Prezzo di vendita": 2.10}
dict6 = {"Quantità": 10, "Prezzo di acquisto": 3.10, "Prezzo di vendita": 6.50}
vegan_store = {"latte di soia": dict1,
               "seitan": dict2, 
               "tofu": dict3, 
               "vegan burger":dict4, 
               "quinoa chips": dict5,
               "vegan mayo":dict6}


def aiuto():
    str_help = f"Elenco comandi:\n1)aggiungi: aggiungi un prodotto al magazzino!\n2)elenca: elenca i prodotti in magazzino\n3)profitti: mostra i profitti totali\n4)chiudi: esci dal programma"
    return print(str_help)

def exit():
    print("bye bye!")
    quit()

def elenca():
    print("Lista Prodotti del Vegan Store\n")
    for prodotto, feature in vegan_store.items():
        print("\nProdotto:", prodotto)
        for chiave in feature:
            print(chiave + ":", feature[chiave])

def aggiungi(prod_name):
    if prod_name in vegan_store.keys():
        quantity_temp =  int(input("Aggiungi una quantità: "))
        quantity = vegan_store[prod_name]["Quantità"]
        prod_buy = vegan_store[prod_name]["Prezzo di acquisto"]
        prod_sell = vegan_store[prod_name]["Prezzo di vendita"]
        spesa_temp = quantity_temp * prod_buy
        spesa.append(spesa_temp)
        new_quantity = quantity + quantity_temp
        dict_temp = {prod_name: {"Quantità": new_quantity, 
                             "Prezzo di acquisto": prod_buy, 
                             "Prezzo di vendita": prod_sell}}
        vegan_store.update(dict_temp)
        print(f"Hai aggiunto {quantity_temp} elementi del prodotto {prod_name}")
        condizione_if = str(input("Vuoi aggiungere un altro prodotto? (si/no)")) 
        if condizione_if == "si":
            prod_name = str(input("Aggiungi un prodotto al magazzino: "))
            aggiungi(prod_name)
        else:
            print("A presto!")
            aiuto()
        
    else:
        quantity =  int(input("Quantità: "))
        prod_sell = float(input("Prezzo di vendita: "))
        prod_buy = float(input("Prezzo di acquisto: "))
        dict_temp = {prod_name: {"Quantità": quantity, 
                             "Prezzo di acquisto": prod_buy, 
                             "Prezzo di vendita": prod_sell}}
        spesa_temp_two = quantity * prod_buy
        spesa.append(spesa_temp_two)
        vegan_store.update(dict_temp)
        print(f"Hai aggiunto {quantity} elementi di un nuovo prodotto --> {prod_name}") 
        condizione_else = str(input("Vuoi aggiungere un altro prodotto? (si/no)"))
        if condizione_else == "si":
            prod_name = str(input("Aggiungi un prodotto al magazzino: "))
            aggiungi(prod_name)
        else:
            print("A presto!")
            aiuto()
        
def vendita(prod_name_sell):
    if prod_name_sell in vegan_store.keys():
        quantity_sell =  int(input("Seleziona una quantità: "))
        quantity = vegan_store[prod_name_sell]["Quantità"]
        prod_buy = vegan_store[prod_name_sell]["Prezzo di acquisto"]
        prod_sell = vegan_store[prod_name_sell]["Prezzo di vendita"]
        new_quantity = quantity - quantity_sell
        ricavo_temp = quantity_sell * prod_sell
        ricavo.append(ricavo_temp)
        dict_temp = {prod_name_sell: {"Quantità": new_quantity, 
                             "Prezzo di acquisto": prod_buy, 
                             "Prezzo di vendita": prod_sell}}
        vegan_store.update(dict_temp)
        print(f"Vendita di {quantity_sell} elementi del prodotto {prod_name_sell} registrata")
        condizione = str(input("Vuoi aggiungere un altro prodotto? (si/no)"))
        if condizione == "si":
            prod_name_sell = str(input("Seleziona un prodotto: "))
            vendita(prod_name_sell)
        else:
            print("A presto!")
            aiuto()
        
    else:
        print(f"Il prodotto selezionato {prod_name_sell}, non è disponibile per il momento nel nostro negozio!")
        prod_name_sell = str(input("Seleziona un altro prodotto: "))
        vendita(prod_name_sell)
        
ricavo = []
spesa = []

while cmd!="chiudi":
    
    cmd = input("Inserisci un comando: ")

    if cmd=="vendita":
        prod_name_sell = str(input("Seleziona un prodotto: "))
        vendita(prod_name_sell)
    elif cmd=="profitti":
        print(f"Il profitto lordo è {round(sum(ricavo),2)} $.\nIl profitto netto è {round((sum(ricavo)-sum(spesa)),2)} $.\nLe spese sono pari a {round(sum(spesa),2)} $.")
    elif cmd=="aggiungi":
        prod_name = str(input("Aggiungi un prodotto al magazzino: "))
        aggiungi(prod_name)
    elif cmd=="elenca":
        elenca()
        
    elif cmd=="aiuto":
        aiuto()
            
    elif cmd=="chiudi":
        exit()
        
    else:
        print("Comando non valido!")
        aiuto()
```
